```python
import warnings
warnings.filterwarnings(action='ignore')
#import requests #selenium을 사용해서 웹 페이지의 내용을 얻어올 것이므로 사용하지 않는다.
from bs4 import BeautifulSoup
from selenium import webdriver
```

```python
#selenium을 사용해 가상 크롬을 실행한다. => 사용중인 크롬 버전(버전 93.0.4577.63)과 같은 버전의 크롬 드라이버를 05에 맞게 다운받는다.
#다운받은 크롬 드라이브를 현재 소스파일이 위치한 경로에 복사해 넣고 사용한다.
driver = webdriver.Chrome('./chromedriver.exe')
#get() 메솓즈로 가상 크롬에 크롤링할 타겟 사이트를 띄운다.
driver.get('https://www.nate.com/')
html = driver.page_source
#print(html)
soup = BeautifulSoup(html, 'html.parser')
print(soup)
```

```python
ranks = soup.findAll('span', {'class' : 'num_rank'})
issues = soup.findAll('a', {'class' : 'ik'})

for i in range(5):
    rank = ranks[i].text
    issue = issues[i].text.strip().split('\n')[0]
    upDown = issues[i].text.strip().split('\n')[1]
    print('{0:>2s}위 : {1}'.format(rank, issue), end='')
    if upDown[:2] == '상승':
        print(' [{}{}]'.format('↑', upDown[2:]))
    elif upDown[:2] == '하락':
        print(' [{}{}]'.format('↓', upDown[2:]))
    elif upDown[:2] == '동일':
        print(' [-]')
    else:
        print(' new')
```
